```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wv
import os

print("Current working directory:", os.getcwd())
```

```python
def lpf(R, C, w, order):
    return (1 / (1 + 1j * w * R * C)) ** order

def hpf(R, C, w, order):
    return ((1j * w * R * C) / (1 + 1j * w * R * C)) ** order
```

```python
R = 100
C = 1e-6

f = np.linspace(20, 20000, 5000)
w = 2 * np.pi * f
```

```python
Hs = 1 / (1 + 1j * w * R * C)
I_order = 20 * np.log(np.abs(Hs))
II_order = 20 * np.log(np.abs(Hs ** 2))

I_phase = np.angle(Hs)
II_phase = np.angle(Hs ** 2)

fig, ax = plt.subplots(2, 1)
ax[0].plot(f, I_order)
ax[0].plot(f, II_order)
ax[0].set_xscale("log")
ax[0].set_ylim(-60, 0)
ax[0].grid(True, which="both")
ax[0].set_xticks([20, 50, 100, 200, 500, 1000, 2000, 5000, 10000, 20000])
ax[0].set_xticklabels(['20', '50', '100', '200', '500', '1k', '2k', '5k', '10k', '20k'])

ax[1].plot(f, I_phase)
ax[1].plot(f, II_phase)
ax[1].set_xscale("log")
ax[1].grid(True, which="both")
ax[1].set_xticks([20, 50, 100, 200, 500, 1000, 2000, 5000, 10000, 20000])
ax[1].set_xticklabels(['20', '50', '100', '200', '500', '1k', '2k', '5k', '10k', '20k'])

plt.show()
```
