```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 1, 1000)
V = 311.126983722 * np.sin(2 * np.pi * 50 * t)
I = 5 * np.sin(2 * np.pi * 50 * t + 20 * np.pi/180)
```

```python
P = V * I

V_rms = np.sqrt(sum(V ** 2) / len(V))
I_rms = np.sqrt(sum(I ** 2) / len(I))
P_avg = sum(P) / len(P)
S = V_rms * I_rms
Q = np.sqrt(S**2 - P_avg**2)
P_fact = P_avg / S

print(f"V RMS          : {V_rms} V")
print(f"I RMS          : {I_rms} A")
print(f"P / Active P   : {P_avg} Watt")
print(f"Q / Reactive P : {Q} VAR")
print(f"S / Apparent P : {S} VA")
print(f"Power Factor   : {P_fact}")

plt.plot(P[:100])
plt.title("Instantanious Power")
plt.grid(True)
plt.show()
```

```python
tm = 24 * 30
W = P_avg * tm

print(f"Energy after {tm} hours : {W / 1000} kWh")
```
